```python
import numpy as np
import pandas as pd
import xlrd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_excel('Train_dataset.xlsx')
df_test=pd.read_excel('Test_dataset.xlsx')
city=df_test["City"]
```

```python
print(df_train.shape,df_test.shape)
```

```python
train_null_values=pd.DataFrame({
    "nullvalues":df_train.isnull().sum(),
    "percentage":100*df_train.isnull().sum()/df_train.shape[0]})
train_null_values.sort_values(by="percentage",ascending=False)
    
```

```python
test_null_values=pd.DataFrame({
    "nullvalues":df_test.isnull().sum(),
    "percentage":100*df_test.isnull().sum()/df_test.shape[0]})
test_null_values.sort_values(by="percentage",ascending=False)
```

```python
df_train.columns
len(df_test[df_test["City"].isin(df_train["City"])==False])
df_test[df_test["Type"].isin(df_train["Type"])==False]
```

```python
df_train.drop(labels=['Popuation [2001]',"City"],axis=1,inplace=True)
df_test.drop(labels=['Popuation [2001]',"City"],axis=1,inplace=True)
```

```python
df_train.head()
```

```python
col=['Population [2011]','Sex Ratio', 'Median Age', 'Avg Temp','Toilets Avl',
       'Water Purity', 'H Index', 'Female Population']
for i in range(len(col)):
    df_train[col[i]].fillna(df_train[col[i]].median(),inplace=True)
    df_test[col[i]].fillna(df_test[col[i]].median(),inplace=True)
    
```

```python
df_train['Foreign Visitors'].fillna(df_train["Foreign Visitors"].mean(),inplace=True)
df_train['Covid Cases'].fillna(df_train['Covid Cases'].mean(),inplace=True)
df_train['# of hospitals'].fillna(df_train['# of hospitals'].mean(),inplace=True)
df_test['# of hospitals'].fillna(0,inplace=True)
df_test['Foreign Visitors'].fillna(0,inplace=True)
df_train["SWM"].fillna("AA",inplace=True)
df_test["SWM"].fillna("AA",inplace=True)
df_test["Type"].fillna("AA",inplace=True)
```

```python
col1=['State', 'Type','SWM']
dict={}
for i in range(len(col1)):
    name=sorted(df_train[col1[i]].unique())
    dict[col1[i]]={c:j for j,c in enumerate(name)}
    df_train[col1[i]].replace(dict[col1[i]],inplace=True)
    df_test[col1[i]].replace(dict[col1[i]],inplace=True)
```

```python
df_train.columns
```

```python
plt.subplot(1,2,1)   
plt.scatter(df_train["Foreign Visitors"],df_train['Covid Cases'])
plt.subplot(1,2,2)
plt.scatter(df_train["Type"],df_train['Covid Cases'])
plt.title("type vs cases")
plt.show()
```

```python
df_train.drop("Type",axis=1,inplace=True)
df_test.drop("Type",axis=1,inplace=True)
```

```python
plt.scatter(df_train["State"],df_train["Covid Cases"])
plt.title("state vs cases")
plt.show()
df_train.drop(labels="State",axis=1,inplace=True)
df_test.drop(labels="State",axis=1,inplace=True)
```

```python
from sklearn.utils import shuffle
df_train=shuffle(df_train)
```

```python
X_train=df_train[['Population [2011]', 'Sex Ratio', 'Median Age',
       'Avg Temp', 'SWM', 'Toilets Avl', 'Water Purity', 'H Index',
       'Female Population', '# of hospitals', 'Foreign Visitors']].values
Y_train=df_train['Covid Cases'].values
X_test=df_test.values
```

```python

scaler_x=preprocessing.StandardScaler().fit(X_train)
X_train=scaler_x.transform(X_train)
X_train, X_val, Y_train, Y_val = train_test_split(X_train,Y_train,test_size=0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Input,BatchNormalization,Dropout
import tensorflow as tf
from keras import metrics,callbacks

```

```python
model=Sequential()
#model.add(tf.keras.layers.Input((11,)))
model.add(Dense(128,input_shape=(11,),activation="relu",kernel_initializer="glorot_uniform",kernel_regularizer="l2"))
model.add(Dropout(0.2))
model.add(Dense(50,activation="relu",kernel_initializer="he_uniform",kernel_regularizer="l1"))
model.add(Dropout(0.2))
model.add(Dense(1,activation="relu"))
```

```python
model.compile(loss="mean_absolute_error",optimizer="adam",metrics=["mape"])
callback=callbacks.EarlyStopping(monitor='val_loss',patience=10)
reduce_lr=callbacks.ReduceLROnPlateau(monitor="val_loss",factor=0.8,patience=0)
hist=model.fit(X_train,Y_train,validation_data=(X_val,Y_val),epochs=1500,batch_size=32,callbacks=[callback,reduce_lr])
```

```python
X_test=scaler_x.fit_transform(X_test)
Y_test=model.predict(X_test)
print(Y_test)
```

```python
Y_test=np.ravel(Y_test)
submission=pd.DataFrame({
    'City':city,
    'Corona_Cases':Y_test
    
})
submission.to_csv("output_01.csv",index=False)
```
